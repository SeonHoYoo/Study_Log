```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs

# 랜덤 시드 고정
np.random.seed(42)

# 1. 데이터 생성 (클러스터와 이상치 포함)
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
outliers = np.random.uniform(low=-8, high=8, size=(20, 2))
X = np.vstack((X, outliers))

# 2. LOF 모델 적용
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)
lof_scores = -lof.negative_outlier_factor_

# 3. 결과 시각화
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='coolwarm', edgecolor='k', s=20)
plt.title('LOF Outlier Detection')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='LOF Score')
plt.show()

# 4. LOF 스코어 데이터프레임 생성
results = pd.DataFrame(X, columns=["Feature_1", "Feature_2"])
results["LOF_Score"] = lof_scores
results["Outlier"] = y_pred
results["Outlier"] = results["Outlier"].apply(lambda x: "Yes" if x == -1 else "No")

# 결과 확인
print(results.sort_values("LOF_Score", ascending=False))
```
