```python
import pandas as pd

user_df = pd.read_csv('data/user.csv')
```

```python
user_df['plus_user'] = user_df['register'].cumsum()
user_df['minus_user'] = user_df['unregister'].cumsum()
user_df['net_user_cumsum'] = user_df['plus_user'] - user_df['minus_user']
user_df.plot(x = 'day', y = 'net_user_cumsum', kind = 'bar')
```
